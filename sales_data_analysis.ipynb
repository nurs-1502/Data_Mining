```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv", index_col = "date", parse_dates = True)
print("Shape of data", df.shape)
df.head()
```

```python
# Converting daily into monthly data
daily_sales_total = df.groupby("date")[["sales"]].sum()
daily_sales_total.head()
```

```python
daily_sales_total.plot(figsize=(12,5))
```

```python
# Take the first difference of the 'sales' column
daily_sales_total_diff = daily_sales_total.diff()
daily_sales_total_diff = daily_sales_total_diff.dropna()
```

```python
daily_sales_total_diff.plot(figsize=(12,5))
```

```python
from statsmodels.tsa.stattools import adfuller

from statsmodels.tsa.stattools import adfuller
def ad_test(dataset):
     dftest = adfuller(dataset, autolag = 'AIC')
     print("1. ADF : ",dftest[0])
     print("2. P-Value : ", dftest[1])
     print("3. Num Of Lags : ", dftest[2])
     print("4. Num Of Observations Used For ADF Regression:",      dftest[3])
     print("5. Critical Values :")
     for key, val in dftest[4].items():
         print("\t",key, ": ", val)
```

```python
ad_test(daily_sales_total_diff)
```

```python
from pmdarima import auto_arima
```

```python
stepwise_fit = auto_arima(daily_sales_total_diff, trace=True, suppress_warnings=True)
stepwise_fit.summary()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
print(daily_sales_total_diff.shape)
train=daily_sales_total_diff.iloc[:-20]
test=daily_sales_total_diff.iloc[-20:]
print(train.shape,test.shape)
```

```python
train.head()
```

```python
model=ARIMA(train,order=(5,0,5))
model=model.fit()
model.summary()
```

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ='levels').rename('ARIMA Predictions')
pred.index = daily_sales_total_diff.index[start:end+1]
print(pred)
```

```python
# Plot both predicted and actual values on the same graph
plt.plot(pred.index, pred.values, label='Predicted')
plt.plot(test.index, test.values, label='Actual')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
test.mean()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rmse=sqrt(mean_squared_error(pred,test))
print(rmse)
```
