```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Load the datasets
train_data = 'Data_Mining-master/train.csv'
test_data = 'Data_Mining-master/test.csv'
train_df = pd.read_csv(train_data)
test_df = pd.read_csv(test_data)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert 'date' columns to datetime
train_df['date'] = pd.to_datetime(train_df['date'])
test_df['date'] = pd.to_datetime(test_df['date'])
```

```python
# Encode the 'family' categorical data
label_encoder = LabelEncoder()
train_df['family_encoded'] = label_encoder.fit_transform(train_df['family'])
test_df['family_encoded'] = label_encoder.transform(test_df['family'])

# Normalize the 'sales' and 'onpromotion' features
sales_scaler = MinMaxScaler()
onpromotion_scaler = MinMaxScaler()
train_df['sales'] = sales_scaler.fit_transform(train_df[['sales']])
train_df['onpromotion'] = onpromotion_scaler.fit_transform(train_df[['onpromotion']])
test_df['onpromotion'] = onpromotion_scaler.transform(test_df[['onpromotion']])
```

```python
# Aggregate data to daily sales per family
daily_train_df = train_df.groupby(['date', 'family_encoded']).agg({
    'sales': 'sum',
    'onpromotion': 'mean'
}).reset_index()
daily_test_df = test_df.groupby(['date', 'family_encoded']).agg({
    'onpromotion': 'mean'
}).reset_index()
```

```python
daily_train_df.head()
```

```python
# Features and target for the model
features = daily_train_df[['date', 'family_encoded', 'onpromotion']]
target = daily_train_df['sales']
```

```python
# Splitting the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Keep the 'date' column separately for error calculation
train_dates = X_train['date']
val_dates = X_val['date']
```

```python
# Drop the 'date' column from features for model training
X_train = X_train.drop(columns=['date'])
X_val = X_val.drop(columns=['date'])
```

```python
# Initialize and train the RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)


# Cross-validate the model
cv_scores = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
cv_rmse_scores = np.sqrt(-cv_scores)

# Make predictions on the test set
test_features = daily_test_df[['family_encoded',  'onpromotion']]
test_predictions = rf_model.predict(test_features)

# Reverse the normalization to get actual sales predictions
predicted_sales = sales_scaler.inverse_transform(test_predictions.reshape(-1, 1))
daily_test_df['predicted_sales'] = predicted_sales

# Output results
print("CV RMSE Scores:", cv_rmse_scores)
print("Average CV RMSE:", cv_rmse_scores.mean())
print(daily_test_df.head())
```

```python
# Make predictions on the validation set
val_predictions = rf_model.predict(X_val)
val_predictions = sales_scaler.inverse_transform(val_predictions.reshape(-1, 1))
y_val_actual = sales_scaler.inverse_transform(y_val.values.reshape(-1, 1))

# Combine dates, actual, and predicted sales
val_results = pd.DataFrame({
    'date': val_dates,
    'actual_sales': y_val_actual.flatten(),
    'predicted_sales': val_predictions.flatten()
})

# Aggregate actual and predicted sales on a daily basis
val_aggregated_daily = val_results.groupby('date').agg({
    'actual_sales': 'sum',
    'predicted_sales': 'sum'
}).reset_index()
```

```python
# Calculate error metrics on the aggregated data
mse = mean_squared_error(val_aggregated_daily['actual_sales'], val_aggregated_daily['predicted_sales'])
rmse = np.sqrt(mse)
mae = mean_absolute_error(val_aggregated_daily['actual_sales'], val_aggregated_daily['predicted_sales'])

# Calculate total actual sales
total_actual_sales = val_aggregated_daily['actual_sales'].sum()

# Calculate percentage error
mse_percentage = (mse / total_actual_sales) * 100
rmse_percentage = (rmse / total_actual_sales) * 100
mae_percentage = (mae / total_actual_sales) * 100

# Print evaluation metrics
print("Validation MSE:", mse)
print("Validation RMSE:", rmse)
print("Validation MAE:", mae)
```

```python
print("Validation MSE Percentage:", mse_percentage)
print("Validation RMSE Percentage:", rmse_percentage)
print("Validation MAE Percentage:", mae_percentage)
```

```python
# plot actual vs predicted values for further insight
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.scatter(val_aggregated_daily['actual_sales'], val_aggregated_daily['predicted_sales'], alpha=0.3)
plt.plot([val_aggregated_daily['actual_sales'].min(), val_aggregated_daily['actual_sales'].max()],
         [val_aggregated_daily['actual_sales'].min(), val_aggregated_daily['actual_sales'].max()],
         'k--', lw=4)  # Perfect prediction line
plt.xlabel('Actual Aggregated Sales')
plt.ylabel('Predicted Aggregated Sales')
plt.title('Validation Predictions vs. Actuals (Aggregated)')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for RandomForestRegressor
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2, 4]

}

# Initialize RandomForestRegressor
rf_model = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)

# Fit the GridSearchCV to find the best parameters
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)

# Use the best parameters to train the RandomForestRegressor
best_rf_model = RandomForestRegressor(**best_params, random_state=42)
best_rf_model.fit(X_train, y_train)

# Cross-validate the model with best parameters
cv_scores = cross_val_score(best_rf_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
cv_rmse_scores = np.sqrt(-cv_scores)

# Make predictions on the test set with the best model
test_predictions = best_rf_model.predict(test_features)

# Reverse the normalization to get actual sales predictions
predicted_sales = sales_scaler.inverse_transform(test_predictions.reshape(-1, 1))
daily_test_df['predicted_sales'] = predicted_sales

# Output results
print("CV RMSE Scores:", cv_rmse_scores)
print("Average CV RMSE:", cv_rmse_scores.mean())
print(daily_test_df.head())
```
